## Binomial  Distribution
<h1>Assumptions of Binomial Distribution </h1>

* **The experiment involves n identical trials.**
* **Each trial has only two possible outcomes denoted as success or as failure.**
*  **Each trial is independent of the previous trials.**
* **The terms p and q remain constant throughout the experiment, where the term p
is the probability of getting a success on any one trial and the term q = (1 - p) is
the probability of getting a failure on any one trial.**

<h2>Formula for binomail distribution</h2>
$$P (x) = nCx\times p^x \times q^{n-x} =\frac {n!}{x!(n - x)!)} \times p^x q^{n-x} $$

where
* n = the number of trials (or the number being sampled)
* x = the number of successes desired
* p = the probability of getting a success in one trial
* q = 1 - p = the probability of getting a failure in one trial

 **$$\mu = np$$**
 **$$\sigma^2 =  npq$$**

exactly 19 are very satisfied with their primary financial institution?**

n=25

x=19

p=0.65

```python
print(binom.pmf(k=19,n=25,p=0.65))
```

**What is the first big change that American drivers made due to higher gas prices?
According to an Access America survey, 30% said that it was cutting recreational
driving. However, 27% said that it was consolidating or reducing errands. If these
figures are true for all American drivers, and if 20 such drivers are randomly sampled
and asked what is the first big change they made due to higher gas prices,**

**a. What is the probability that exactly 8 said that it was consolidating or reducing errands?**

**b. What is the probability that none of them said that it was cutting recreational
driving?**

**c. What is the probability that more than 7 said that it was cutting recreational
driving?**

```python
#a. What is the probability that exactly 8 said that it was consolidating or reducing errands?
print(binom.pmf(k=8,n=20,p=0.27))
```

```python
#b. What is the probability that none of them said that it was cutting recreational driving?
print(round(binom.pmf(k=0,n=20,p=0.30),5))
```

```python
#c. What is the probability that more than 7 said that it was cutting recreational driving?
prob=1-binom.cdf(7,20,0.30)
print(round(prob,3))
```
